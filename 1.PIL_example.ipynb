```python
from PIL import Image

import matplotlib.pyplot as plt
```

```python
## image read
img = Image.open('./data/1.PIL-image001.jpeg')


## imshow()는 데이터를 그림에 그리고 show()는 실제로 이 그림을 화면에 표시
## image visualize
plt.imshow(img)     ## 이미지나 데이터배열을 시각화하는데 사용
                    ## 화면에 표시만하고 실제 창안띄움.

plt.show()          ## 그림 그림을 화면에 표시할때사용.
                    ## 실제 창에 표시할때 사용
img
```

## Image resize

```python
w,h = img.size
img_resize = img.resize((w//2, h//2))   ## 반드시 integer값을 넣어줘야한다.
plt.imshow(img_resize)
plt.show()

img_resize = img.resize((w*2, h*2))
plt.imshow(img_resize)
plt.show()


## resize options -> 
# https://pillow.readthedocs.io/en/stable/handbook/concepts.html?highlight=resize#PIL.Image.NEAREST
```

## Image crop

```python

def two_imge_show(im1,im2):
    plt.subplot()

def two_image_show(im1, im2):
    plt.subplot(1,2,1)      ## (nrows, ncols, index)
    plt.imshow(im1)
    plt.subplot(1,2,2)    
    plt.imshow(im2)
    plt.show()
    
img_crop = img.crop((w//4, h//4, w*3//4, h*3//4))   ## crop한 위치좌표만 넣어주어도 된다.

plt.imshow(img_crop)
## (left, upper, right, lower)

two_image_show(img, img_crop)
```

## Image rotate

```python
img_rotate = img.rotate(45)   ## degree, NOT radian

#img_rotate.show() 
plt.imshow(img_rotate)
```

```python
img_flip = img.transpose(Image.FLIP_LEFT_RIGHT)   ## FLIP_TOP_BOTTOM

two_image_show(img, img_flip)
```

```python
img_flip_and_rotate = img.rotate(45).transpose(Image.FLIP_LEFT_RIGHT)

img_flip_and_rotate.save('save_img.png')
```

# 결론

이미지 처리 그 자체만으로는 PIL 라이브러리가 훨씬 간단하고 편리하다.

하지만, 복잡한 컴퓨터비전 알고리즘의 적용이나, 딥러닝/머신러닝 모델을 개발할 때에는 PIL 의 함수만으로는 수행할 수 없다.

그럴때는 tensorflow나 pytorch를 사용해야 한다.
